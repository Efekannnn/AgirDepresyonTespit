```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"KullanÄ±lan cihaz: {device}")

df = pd.read_csv('cleaned_combined_data.csv')

val_texts = df['cleaned_text'].tolist()
val_labels = df['label'].tolist()

from google.colab import drive
drive.mount('/content/drive')

model_name = "/content/drive/MyDrive/my_mental_health_model/roberta"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.to(device)

val_encodings = tokenizer(val_texts, truncation=True, padding=True, return_tensors='pt')

batch_size = 32
predicted_labels = []

with torch.no_grad():
    for i in range(0, len(val_encodings['input_ids']), batch_size):
        input_ids_batch = torch.tensor(val_encodings['input_ids'][i:i + batch_size]).to(model.device)
        attention_mask_batch = torch.tensor(val_encodings['attention_mask'][i:i + batch_size]).to(model.device)
        logits = model(input_ids=input_ids_batch, attention_mask=attention_mask_batch).logits
        predicted_labels.extend(torch.argmax(logits, dim=1).tolist())

dogruluk = accuracy_score(val_labels, predicted_labels)
kesinlik = precision_score(val_labels, predicted_labels)
geri = recall_score(val_labels, predicted_labels)
f1 = f1_score(val_labels, predicted_labels)

tn, fp, fn, tp = confusion_matrix(val_labels, predicted_labels).ravel()

ozgulluk = tn / (tn + fp)
hassaslik = tp / (tp + fn)

auc = roc_auc_score(val_labels, predicted_labels)

print(f"Accuracy: {dogruluk}")
print(f"Precision: {kesinlik}")
print(f"Recall: {geri}")
print(f"F1-Score: {f1}")
print(f"Specificity: {ozgulluk}")
print(f"Sensitivity: {hassaslik}")
print(f"AUC: {auc}")

with open("/content/drive/MyDrive/my_mental_health_model/electra/metrics_roberta.txt", "w") as f:
    f.write(f"Accuracy: {dogruluk}\n")
    f.write(f"Precision: {kesinlik}\n")
    f.write(f"Recall: {geri}\n")
    f.write(f"F1-Score: {f1}\n")
    f.write(f"Specificity: {ozgulluk}\n")
    f.write(f"Sensitivity: {hassaslik}\n")
    f.write(f"AUC: {auc}\n")

metrics = {
    "Accuracy": dogruluk,
    "Precision": kesinlik,
    "Recall": geri,
    "F1-Score": f1,
    "Specificity": ozgulluk,
    "Sensitivity": hassaslik,
    "AUC": auc
}

plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values())
plt.xlabel("Metrikler")
plt.ylabel("DeÄerler")
plt.title("Model PerformansÄ± Metrikleri")
plt.ylim([0, 1])
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


```
